#### Operaciones básicas

```Matplotlib``` es la API de ```pyplot``` predeterminada y tiene una interfaz global con un estilo muy similar a MATLAB

````python
# Creating an array 'x' with values ranging from -10 to 10 using numpy's array function
x = np.array(range(-10, 11))
x
````

````python
# Creating a new figure for the plot with a specified figure size (8x4 inches)
plt.figure(figsize=(8, 4))

# Adding a title to the plot
plt.title('My Nice Plot')

# Plotting the first curve, which is x squared (x^2)
# The curve is colored green, labeled as 'x^2', and the points are marked with circles ('o')
plt.plot(x, x**2, color='green', label='$x^2$', marker='o')

# Plotting the second curve, which is the negative of x squared (-x^2)
# The curve is colored red, labeled as '-x^2', and the points are marked with crosses ('x')
plt.plot(x, -1 * (x ** 2), color='r', label='$-x^2$', marker='x')

# Labeling the x-axis with the LaTeX-style variable name 'x'
plt.xlabel('$x$ - axis')

# Labeling the y-axis with the LaTeX-style variable name 'y'
plt.ylabel('$y$ - axis')

# Adding a legend to the plot to distinguish the two curves
plt.legend()

# Adding a grid to the plot for better readability
plt.grid()

# Displaying the plot
plt.show()
````

Podemos mejorar el estilo de los títulos agregando los parámetros de la fuente como un diccionario

````python
font = {'family':'serif','color':'blue','size':20}

plt.figure(figsize=(8, 4))
plt.title('My Nice Plot', fontdict=font)
plt.plot(x, x ** 2, label= "$x^2$", color='c', marker='')
plt.plot(x, -1 * (x ** 2), label='$-x^2$', color='m', marker='')
plt.xlabel('$x$', fontdict=font)
plt.ylabel('$y=x^2$', fontdict=font)
plt.legend()
plt.grid()
plt.show()
````

Es posible generar varias gráficas de forma simultanea utilizando el comando ```plt.subplot(rows,columns, panel_selected)```

````python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)  # rows, columns, panel selected
plt.plot(x, x ** 2)
plt.plot([0, 0, 0], [-10, 0, 100])
plt.legend(['$x^2$', 'Vertical Line'])
plt.xlabel('$x$')
plt.ylabel('$y=x^2$')
plt.grid()
plt.title('A nice plot')

plt.subplot(1, 2, 2)  # rows, columns, panel selected
plt.plot(x, -1 * (x ** 2))
plt.plot([-10, 0, 10], [-50, -50, -50], color='green')
plt.legend(['$-x^2$', 'Horizontal Line'])
plt.xlabel('$x$')
plt.ylabel('$y=-x^2$')
plt.title('Another nice plot')
plt.grid()

plt.show()
````

````python
plt.figure(figsize=(18, 9))

equation_names = {
    1: "Linear Equation",
    2: "Quadratic Equation",
    3: "Cubic Equation",
    4: "Quartic Equation",
    5: "Quintic Equation",
    6: "Sextic Equation"
}

for i in range(6):
    plt.subplot(2, 3, i + 1)
    y = x ** (i+1)
    plt.plot(x, y)
    plt.plot([0, 0, 0], [min(y), 0, max(y)])
    plt.xlabel('$x$')
    plt.ylabel(f'$y = x^{i+1}$')
    plt.legend([f'$x^{i+1}$', 'Vertical Line'])
    plt.grid()
    plt.title(equation_names[i + 1])
````

Se pueden generar de manera automática un conjunto de gráficas utilizando el objeto axes y un ciclo simple


````python
import random as r

rows = 3
columns = 3
fig, axes = plt.subplots(rows, columns, figsize = (20,14))
fig.suptitle('Linear functions', size=28, weight=25, color='r')

for row in range(rows):
    for column in range(columns):
        slope = random.randint(-5, 5)
        intercept = round(r.random(), 1)
        my_figure = axes[row][column]
        my_figure.plot(x, slope * x + intercept, label = f'$y = {slope}x + {intercept}$', color = np.random.rand(3))
        my_figure.set_xlabel('$x$')
        my_figure.set_ylabel('$y$')
        my_figure.set_title(f'Figure ({row},{column})')
        my_figure.grid()
        my_figure.legend()
````

```Matplotlib``` tiene una gran cantidad de mapas de colores. Un mapa de colores es una especie de lista de colores en la cual cada color está ubicado en un rango entre ```1``` y ```100```.
Se pueden usar estos mapas para configurar un color variable en nuestros gráficos. Adicionalmente, si queremos agregar la barra de colores podemos utilizar el comando ```plt.colorbar()```

````python
n = 50
colors = np.random.randint(0, 100, size=n)

plt.figure(figsize=(8, 4))
plt.scatter(x1, y1, c=colors, marker='o', s=20, cmap='viridis')
plt.title('A fancy scatter plot using color maps')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar()
plt.grid()
````

Es posible realizar configuraciones más elaboradas especificando valores variables para el área de cada punto con el atributo ```s``` y así como el nivel de trasparencia con el atributo ```alpha```.

````python
plt.figure(figsize=(8, 4))
area = np.pi * (20 * np.random.rand(N))**2  # 0 to 15 point radii
plt.scatter(x1, y1, s=area, c=colors, alpha=0.5, cmap='Dark2')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar()
plt.show()
````

#### Histogramas

Otro gráfico de utilidad son los histogramas. Estos nos permiten comprobar la distribución de frecuencias de un conjunto de datos. El siguiente ejemplo muestra un histograma construido a partir de una muestra aleatoria de una distribución normal. El parámetro ```bins``` nos permite ajustar el valor de los intervalos para el histograma

````python
values = np.random.randn(10000)
plt.subplots(figsize=(8, 4))
plt.hist(values, bins='auto', alpha=0.8, histtype='bar', color='steelblue', edgecolor='green')
plt.xlim(xmin=-5, xmax=5)
plt.show()
````

Podemos utilizar scipy para estimar la función de densidad de la muestra aleatoria mediate el método ```stats.kde.gaussian_kde()```


````python
from scipy.stats import gaussian_kde
density = gaussian_kde(values)
````

````python
plt.subplots(figsize=(8, 4))

values2 = np.linspace(min(values) - 5, max(values) + 5, 100)

plt.plot(values2, density(values2), color='orange')
plt.fill_between(values2, 0, density(values2), alpha=0.1, color='orange')
plt.xlim(xmin=-5, xmax=5)

plt.show()
````

Finalmente, es posible comparar el histograma con la función de densidad estimada.

````python
plt.subplots(figsize=(8, 4))

plt.hist(values, bins='auto', alpha=0.8, density=True, histtype='bar', color='steelblue', edgecolor='green')
plt.plot(values2, density(values2), color='orange', linewidth=1.0)

plt.xlim(xmin=-5, xmax=5)
plt.show()
````

#### Diagrama de cajas


Los diagramas de cajas nos permiten representar un conjunto de datos a partir de sus cuartiles. Son especialmente útiles para visualizar posibles datos atípicos.

````python
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(80, 30, 200)
data3 = np.random.exponential(100, 200)
data4 = np.random.normal(70, 25, 200)

data_to_plot = [data1, data2, data3, data4]
plt.figure(figsize=(8, 4))
plt.boxplot(data_to_plot)
plt.show()
````

### Exploración de datos con Pandas

Pandas ofrece ciertos métodos de visualización que nos permiten explorar un DataFrame. A continuación se presenta algunos ejemplos de como acceder a dichas funcionalidade utilizando un dataset con información sobre el ingreso promedio en diferentes profesiones

````python
df.plot(x="Rank", y=["P25th", "Median", "P75th"], figsize=(8,4), grid=True)
plt.show()
````

El método ```plot()``` en Pandas tiene diferentes parámetros de configuración. Uno de los más importantes el atributo ```kind``` que nos permite seleccionar entre diferentes tipos de gráficos:
+ ``` area ```
+ ``` bar ```
+ ``` barh ```
+ ``` box ```
+ ``` hexbin ```
+ ``` hist ```
+ ``` kde ```
+ ``` line ```
+ ``` pie ```
+ ``` scartter ```

El valor por defecto para el atributo ```kind``` es el tipo de gráfico line. Adicionalmente podemos utilizar muchos de los atributos de configuración que usamos en ```matplotlib```

````python
df.plot(x="Rank", y="P75th", figsize=(14,6), grid=True, color='r', marker='o', linestyle="dashed", title="A panda line plot",)
plt.show()
````
