##### Functions

```python
# Pandas Importing
import pandas as pd 

def get_file_details(file):

    # Reading the CSV file into a DataFrame, with the specified column set as the index.
    df = pd.read_csv(file)  

    # Stripping leading and trailing spaces from the column headers.
    df.columns = df.columns.str.strip()

    # Returning the cleaned DataFrame.
    return df 
```

```python
#Filtering and grouping alcohol consumption by school and sex.
grouped_df = df.groupby(['school', 'sex']).size().unstack(fill_value=0)
grouped_df
```

Graph of filtering and grouping of alcohol consumption by school and sex.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Configure the graph
fig, ax = plt.subplots(figsize=(8, 6))

# Define the width of the bars
bar_width = 0.35

# Positions for the bars on the x axis
index = np.arange(len(grouped_df))

# Get the column values, with a default value of 0 if they do not exist
male_counts = grouped_df.get('M', pd.Series(0, index=grouped_df.index))
female_counts = grouped_df.get('F', pd.Series(0, index=grouped_df.index))

# Add tags and title
bar1 = ax.bar(index, male_counts, bar_width, label='Male', alpha=0.5)
bar2 = ax.bar(index + bar_width, female_counts, bar_width, label='Female', alpha=0.5)

# Add tags and title
ax.set_xlabel('School')
ax.set_ylabel('Count')
ax.set_title('Count of Records by School and Sex')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(grouped_df.index)
ax.legend()

# Show
plt.show()
```

### ‚úçÔ∏è Conclusion

The data reflect alcohol consumption among students at two schools, GP and MS. At both schools, female students report higher alcohol consumption than male students. Specifically, at GP, 420 cases of alcohol consumption were recorded among female students, compared to 352 cases of alcohol consumption among male students. Similarly, at MS, female students reported 171 cases of alcohol consumption, while male students reported 101. This suggests that alcohol consumption is higher among female students at both schools.

This relationship can be seen in the fact that there are more female students enrolled at both schools than male students and that GP has a larger student population than MS.s.

```python
grouped_df  = df5.groupby(['school', 'age']).size().reset_index(name='count')
gp_data = grouped_df[['age', 'count']][grouped_df['school'] == 'GP']
ms_data = grouped_df[['age', 'count']][grouped_df['school'] == 'MS']
```

```python
# Configurar el gr√°fico de l√≠neas
plt.figure(figsize=(10, 6))

# Graficar la primera l√≠nea para la escuela 'GP'
plt.plot(gp_data['age'], gp_data['count'], marker='o', linestyle='-', label='GP', color='blue')

# Graficar la segunda l√≠nea para la escuela 'MS'
plt.plot(ms_data['age'], ms_data['count'], marker='o', linestyle='-', label='MS', color='green')

# A√±adir etiquetas y t√≠tulo
plt.xlabel('Age')
plt.ylabel('Number of Students')
plt.title('Alcohol Consumption by Age and School')
plt.legend()

# Mostrar el gr√°fico
plt.grid(True)
plt.show()
```

### ‚úçÔ∏è Conclusion

The data indicates that students aged 15 to 18 in both schools exhibit high levels of alcohol consumption. However, this behavior starts to decline after the age of 19.
This trend may be attributed to the developmental stage of adolescence, during which peer influence and social experimentation often lead to increased alcohol consumption. As students transition into young adulthood, they may develop a greater sense of responsibility and face new life challenges that can contribute to reduced drinking behavior.

```python
#Filtering Alcohol Consumption by Criteria: age
grouped_df  = df5.groupby(['age']).size().reset_index(name='count')
grouped_df.sort_values(by='count', ascending=False)
```

```python
plt.figure(figsize=(8, 8))
plt.pie(grouped_df['count'], labels=grouped_df['age'], autopct='%1.1f%%')
plt.title('Alcohol Consumption by Age (GP)')
plt.legend(grouped_df['age'], title='Ages', loc='upper left')
plt.show()
```

### ‚úçÔ∏è Conclusion

The data reveals a notable pattern in alcohol consumption among students across different ages. The highest levels of consumption are observed in students aged 15 to 18, with a significant decrease in consumption as students age. Specifically:

+ Students aged 15 have the highest consumption with a count of 194.
+ Consumption remains high for students aged 16 (281) and 17 (277), and slightly drops for students aged 18 (222).
+ A sharp decline is evident from age 19 onward, with consumption dropping to 56 for age 19, and further decreasing to minimal levels by ages 20 (9), 21 (3), and 22 (2).

This trend suggests that alcohol consumption among students peaks during mid-adolescence and decreases significantly as they transition into young adulthood.
