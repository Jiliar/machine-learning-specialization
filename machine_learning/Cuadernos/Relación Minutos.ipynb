#### **‚≠ïÔ∏è Variables con datos faltantes, Porcentaje total de valores faltantes en la base de datos.**

```python
import json

def valores_faltantes_por_columna(df):
    # Escribir las variables que tienen datos faltantes, adem√°s, proporcionar el porcentaje total de valores faltantes en la base de datos.
    # Paso 1: Encontrar columnas con valores faltantes
    faltantes_por_columna = df.isnull().sum()
    
    # Paso 2: Filtrar columnas que tienen valores faltantes
    data1 = faltantes_por_columna[faltantes_por_columna > 0]
    variables_con_faltantes = data1.to_dict()
    
    print('\nCant. valores faltantes por columna:',
          json.dumps(variables_con_faltantes, indent=4),
          sep="\n")
    
    # Paso 3: Calcular el porcentaje de valores faltantes por columna
    data = (data1 / len(df)) * 100
    data = data.to_dict()
    porcentaje_faltantes_por_columna = {key: f'{value:.2f}%' for key, value in data.items()}
    
    # Paso 4: Calcular el porcentaje total de valores faltantes en el DataFrame
    porcentaje_total_faltantes = (df.isnull().sum().sum() / df.size) * 100
    porcentaje_total_no_faltantes = (df.notnull().sum().sum() / df.size) * 100
    
    # Resultados
    print('\n% valores faltantes por columna:', f'{json.dumps(porcentaje_faltantes_por_columna, indent=4)}', sep="\n")
    
    print(f"\nPorcentaje total de valores faltantes en la base de datos: {porcentaje_total_faltantes:.2f}%")
    print(f"Porcentaje total de valores en la base de datos: {porcentaje_total_no_faltantes:.2f}%")

valores_faltantes_por_columna(df)
```

#### **‚≠ïÔ∏è Tratamiento de valores nulos**

```python
# Imputaci√≥n por K-Nearest Neighbors (KNN)
from sklearn.impute import KNNImputer  # Importar el KNNImputer de la biblioteca sklearn

# Crear una copia del DataFrame original
df_knn = df

# Reemplazar las categor√≠as de la columna 'creditcd' de 'Y' y 'N' a 1 y 0 respectivamente
df_knn['creditcd'] = df['creditcd'].replace({'Y': 1, 'N': 0})

# Inicializar el imputador KNN con 5 vecinos
imputer = KNNImputer(n_neighbors=5)

# Aplicar el imputador a la columna 'creditcd'
# Esto transforma los valores faltantes en base a la informaci√≥n de los 5 vecinos m√°s cercanos
df_knn[['creditcd']] = imputer.fit_transform(df_knn[['creditcd']])

# Convertir los valores continuos imputados a binarios (1 o 0)
# Si el valor es mayor que 0.5, se asigna 1, de lo contrario se asigna 0
df_knn['creditcd'] = df_knn['creditcd'].apply(lambda x: 1 if x > 0.5 else 0)

# Obtener los valores √∫nicos de la columna 'creditcd' para verificar los resultados
values = df_knn['creditcd'].unique()
print(f'\nK-Nearest Neighbors (KNN):')
print(f'Cantidad de Valores: {str(len(values))}')  # Imprimir la cantidad de valores √∫nicos
print(f'Valores: {sorted(values)}')  # Imprimir los valores √∫nicos ordenados

# Llamar a una funci√≥n para mostrar los valores faltantes por columna (debe estar definida en otro lugar)
valores_faltantes_por_columna(df_knn)
df['creditcd'] = df['creditcd'].astype(object)
```

#### **üìà Grafico de barras apiladas**

```python
import warnings
import statsmodels.api as sm

# Ignorar advertencias
warnings.filterwarnings("ignore")


# Supongamos que df_knn ya est√° definido y tiene las columnas 'totmou' y 'totrev'
# Primero, crea bins para totmou
bins = [0, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]
labels = ['0-100', '101-200', '201-300', '301-400', '401-500', 
          '501-600', '601-700', '701-800', '801-900', '901-1000']

df_knn['totmou_bins'] = pd.cut(df_knn['totmou'], bins=bins, labels=labels)

# Calcular el total para cada bin
totals = df_knn.groupby('totmou_bins')[['totmou', 'totrev']].sum().reset_index()

# Crear gr√°fico apilado de barras
plt.figure(figsize=(9, 5))

# Configurar el gr√°fico apilado
plt.bar(totals['totmou_bins'], totals['totmou'], label='Total Minutos Usados (totmou)', color='blue')
plt.bar(totals['totmou_bins'], totals['totrev'], bottom=totals['totmou'], label='Total Ingresos (totrev)', color='green')

# A√±adir t√≠tulos y etiquetas
plt.title('Gr√°fico Apilado de Barras: Minutos Usados y Ingresos Totales', fontsize=10)
plt.xlabel('Rango de Minutos Usados', fontsize=10)
plt.ylabel('Total', fontsize=10)

for i in range(len(totals)):
    plt.text(i, totals['totmou'][i] / 2, f"{totals['totmou'][i]:.0f}", ha='center', fontsize=6, color='white')
    plt.text(i, totals['totmou'][i] + totals['totrev'][i] / 2, f"{totals['totrev'][i]:.0f}", ha='center', fontsize=6, color='white')

plt.xticks(rotation=45, fontsize=8)  # Cambiar el tama√±o de las etiquetas en el eje X
plt.yticks(fontsize=8)  # Cambiar el tama√±o de las etiquetas en el eje Y

# Mostrar leyenda
plt.legend(fontsize=8)

# Mostrar el gr√°fico
plt.tight_layout()  # Ajustar el layout
plt.show()

df_knn = df_knn.drop(columns=['totmou_bins'])
```

#### **‚≠ïÔ∏è Conclusi√≥n Grafico de barras apiladas para describir relaci√≥n entre Minutos v.s Ingreso de los Clientes** 

En el grafico de barras apiladas dado un rango de minutos usados prestablecido, se puede ver que hay cierta proporcionalidad en el comportamiento de las variables minutos usados e Ingresos, a medida que una crece tambien lo hace la otra; posiblemente hay una correlaci√≥n fuerte entre ambas variables.

#### **‚≠ïÔ∏è Graficos para describir relaci√≥n entre Minutos v.s Ingreso de los Clientes** 

```python
# Seleccionar las columnas relevantes
X = df_knn['totmou']  # Variable independiente (total de minutos usados)
y = df_knn['totrev']  # Variable dependiente (ingresos)

# Agregar una constante a la variable independiente para el t√©rmino de intercepto
X = sm.add_constant(X)

# Ajustar el modelo de regresi√≥n lineal
model = sm.OLS(y, X).fit()

# Resumen de los resultados del modelo
print(model.summary())

# Extraer los par√°metros estimados utilizando .iloc
intercept = model.params.iloc[0]  # Œ± (intercepto)
slope = model.params.iloc[1]  # Œ≤ (pendiente)

# Mostrar los par√°metros estimados
print(f'Intercepto (Œ±): {intercept:.2f}')
print(f'Pendiente (Œ≤): {slope:.2f}')

# Visualizar la relaci√≥n con la l√≠nea de regresi√≥n
# Gr√°fico con l√≠nea de regresi√≥n, medias y etiquetas de la funci√≥n
plt.figure(figsize=(10, 6))

# Gr√°fico de dispersi√≥n
plt.scatter(df_knn['totmou'], df_knn['totrev'], alpha=0.5, label='Datos Observados')

# T√≠tulo y ejes
plt.title('Regresi√≥n Lineal: Minutos Usados vs Ingresos')
plt.xlabel('Total de Minutos Usados')
plt.ylabel('Ingresos Totales (USD)')

# L√≠neas de promedio
plt.axhline(y=df_knn['totrev'].mean(), color='magenta', linestyle='--', label='Promedio de Ingresos')
plt.axvline(x=df_knn['totmou'].mean(), color='g', linestyle='--', label='Promedio de Minutos Usados')

# L√≠nea de regresi√≥n
plt.plot(df_knn['totmou'], intercept + slope * df_knn['totmou'], color='red', label='L√≠nea de Regresi√≥n')

# Mostrar la ecuaci√≥n en la parte superior derecha del gr√°fico
equation_text = f'totrev = {slope:.2f} * totmou + {intercept:.2f}'
plt.text(df_knn['totmou'].max(), df_knn['totrev'].max(), equation_text, fontsize=10, color='black',
         verticalalignment='top', horizontalalignment='right',
         bbox=dict(facecolor='white', edgecolor='gray', boxstyle='round,pad=0.5'))

# Leyenda y cuadr√≠cula
plt.legend(loc='upper left')
plt.grid()

# Mostrar gr√°fico
plt.show()

# Definici√≥n de la funci√≥n de regresi√≥n
def predict_totrev(totmou):
    return slope * totmou + intercept

# Ejemplo de uso de la funci√≥n predict_totrev
# Calcular los ingresos para diferentes valores de minutos usados
minutos_usados = []
for i in range(0, 1001, 100):  # De 100 a 1000 minutos en incrementos de 100
    minutos_usados.append(i)
ingresos = [predict_totrev(totmou) for totmou in minutos_usados]

# Imprimir los resultados
for totmou, rev in zip(minutos_usados, ingresos):
    print(f"Minutos usados: {totmou} minutos, Ingresos: {rev:.2f}")
```

#### **‚≠ïÔ∏è Conclusi√≥n relaci√≥n entre Minutos v.s Ingreso de los Clientes**

El modelo de regresi√≥n lineal muestra que existe una relaci√≥n positiva significativa entre los minutos utilizados por los clientes (`totmou`) y los ingresos totales generados (`totrev`). El R-cuadrado de `0.698` indica que el `69.8%` de la variabilidad en los ingresos puede explicarse por los minutos usados. La pendiente `(Œ≤ = 0.08)` sugiere que, por cada minuto adicional usado, los ingresos aumentan en promedio en `0.08` unidades. El intercepto (`Œ± = 405.83`) representa los ingresos estimados cuando los minutos utilizados son cero.

El gr√°fico de dispersi√≥n muestra una concentraci√≥n significativa de datos alrededor del punto de intersecci√≥n entre las l√≠neas de promedio de ingresos, promedio de minutos utilizados y la l√≠nea de regresi√≥n totrev = 0.08 * totmou + 405.83, lo que indica que la mayor√≠a de los clientes se sit√∫an cerca de estos valores medios. A medida que aumenta la pendiente de la l√≠nea de regresi√≥n, algunos datos se dispersan, reflejando que, aunque existe una clara relaci√≥n positiva entre los minutos usados y los ingresos, no todos los clientes siguen esta tendencia de manera uniforme, sugiriendo la influencia de otros factores en algunos casos.
